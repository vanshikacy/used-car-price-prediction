```python
import pandas as pd
df=pd.read_csv("./data/car_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Selling_Price"], bins=30)
plt.xlabel("Selling Price (in lakhs)")
plt.ylabel("Number of Cars")
plt.title("Distribution of Selling Price")
plt.show()
```

```python
current_year=2024
df["car_age"]=current_year-df["Year"]
```

```python
df[["Year", "car_age"]].head()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["car_age"], df["Selling_Price"], alpha=0.6)
plt.xlabel("Car Age (years)")
plt.ylabel("Selling Price (in lakhs)")
plt.title("Selling Price vs Car Age")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Kms_Driven"], df["Selling_Price"], alpha=0.6)
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price (in lakhs)")
plt.title("Selling Price vs Kms Driven")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
df.boxplot(column="Selling_Price", by="Owner")
plt.xlabel("Number of Owners")
plt.ylabel("Selling Price (in lakhs)")
plt.title("Selling Price vs Number of Owners")
plt.suptitle("")
plt.show()
```

```python
df["company"].value_counts()
```

```python
top_brands=df["company"].value_counts().head(8).index
df_top=df[df["company"].isin(top_brands)]
```

```python
plt.figure(figsize=(10, 6))
df_top.boxplot(column="Selling_Price", by="company")
plt.xlabel("Company")
plt.ylabel("Selling Price (in lakhs)")
plt.title("Selling Price Distribution by Company")
plt.suptitle("")
plt.xticks(rotation=45)
plt.show()
```
